# Logistic Regression With Gradient Descent


```python
from math import exp
import sys

def sigmoid(x):
    try:
    
        sig= exp(x) / (1 + exp(x))

    except OverflowError as err:
        if (x>0):
            return 1
        else: 
            return 0
    
    return sig
```

```python
def linearCombination(x, w):
    sum = 0
    for i in range(len(w)):
        sum += w[i] * x[i]
    return sum
```

```python
def p1(x, w):
    return sigmoid(linearCombination(w,x))
```

```python
from numpy import zeros,append

def gradient(X, y, w, lam):
    sum = zeros(len(w))
    for i in range(len(X)):
        for j in range(len(w)):
            sum[j] += (p1(X[i],w) - y[i]) * X[i][j] 
            
    for j in range(len(w)):
        sum[j] += lam * w[j]
        
    return sum
```

```python
import numpy as np
import matplotlib.pyplot as plt
def gradientDescent(X,y,lr,lam,maxIt):
    global w0, w1, w2
    w0 = []
    w1 = []
    w2 = []
    w = zeros(len(X[0]))
    for gen in range(maxIt):
        grad = gradient(X,y,w,lam);
        for i in range(len(w)):
            w[i] = w[i] - lr * grad[i]
        print(w)
        w0= append(w0,w[0])
        w1= append(w1,w[1])
        w2= append(w2,w[2])
    return w
```

# 2  Newton-Raphson Weight Update Rule

```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False #用来正常显示负号

x=np.arange(-2,2,0.01)
y=x*x*x*x-5*x-3
plt.xlabel('x')
plt.ylabel('y')
plt.title("Image of x^4-5x-3")
plt.plot(x, y)
plt.show()
```

```python
def derivative(x):
    return 4*(x**3)-5
def twoderivative(x):
    return 12*(x**2)

def findPoint(x):
    next=0
    
#next是什么意思
    if derivative(x)==0:
        return x
    else:
        next=x-derivative(x)/twoderivative(x)
    if abs(derivative(x)-derivative(next))<1e-6:
        print('when x='+str(next)+' the problem reach the optimum value ')
    else:
        return findPoint(next)
    
findPoint(-6)
```
